```python
print(df.shape)
df=df.dropna(axis=0, how='any')
print(df.shape)
```

```python
result = df.loc[:,'id']
result = pd.DataFrame()
print(result.shape)
```

```python
for i in df.columns:
    print('-------------------'+str(i)+'-------------------')
    if str(i)=='id':
        continue
    catenc = pd.factorize(df.loc[:,i])
    Y2 = pd.DataFrame(data=catenc[0],columns=[str(i)])
    print(result.shape)
    result = pd.concat([result, Y2], axis=1, sort=False)
    print(result.shape)
```

## Resampling of Data (UpSample) - Balancing

```python
Y = result.loc[:,'target']
X = result.drop(columns=['target','day','month'])
print(X.shape)
print(Y.shape)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE('minority')
X_rs, Y_rs = sm.fit_sample(X, Y)
```

```python
print(X_rs.shape)
print(Y_rs.shape)
```

```python
Y_rs.value_counts()
```

```python
X_rs.to_csv('X_resampled.csv', index=False, encoding='utf-8-sig')
Y_rs.to_csv( "Y_resampled.csv", index=False, encoding='utf-8-sig')
```
